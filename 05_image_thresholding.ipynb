```python
#import necessary packeages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
#display image with method
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
#load image with method
def load_image():
    image = cv2.imread('images/girl.jpg')
    return image
```

```python
load_img = load_image()
display_img(load_img)
```

```python
#display gray image
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
display_img(load_img, cmap="gray")
```

```python
#binary thresholding
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
ret, thes1= cv2.threshold(load_img,127,255,cv2.THRESH_BINARY)
display_img(thes1, cmap ="gray")
```

```python
#inverse binary thresholding
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
ret, thes2= cv2.threshold(load_img,127,255,cv2.THRESH_BINARY_INV)
display_img(thes1, cmap ="gray")
```

```python
#load bgr2rgb image
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2RGB)
#ret, thes1= cv2.threshold(load_img,127,255,cv2.THRESH_BINARY_INV)
display_img(load_img)
```

```python
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
ret, thes3= cv2.threshold(load_img,127,255,cv2.THRESH_TOZERO)
display_img(thes3, cmap ="gray")
```

```python
#threshold truncation
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
ret, thes4= cv2.threshold(load_img,127,255,cv2.THRESH_TRUNC)
display_img(thes4, cmap ="gray")
```

```python
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
#ret, thes4= cv2.threshold(load_img,127,255,cv2.THRESH_TRUNC)

th5 = cv2.adaptiveThreshold(load_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,8)
display_img(thes4, cmap ="gray")
```

```python
#adding weight to images 
load_img = load_image()
load_img = cv2.cvtColor(load_img, cv2.COLOR_BGR2GRAY)
#ret, thes4= cv2.threshold(load_img,127,255,cv2.THRESH_TRUNC)
#th5 = cv2.adaptiveThreshold(load_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,8)
th6 = cv2.addWeighted(src1 = load_img, alpha = 0.7, src2 = th5, beta = 0.3, gamma = 0)
display_img(th6, cmap ="gray")
```
