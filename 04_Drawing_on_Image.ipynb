```python
#import necessary packeages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
#create a blnk image
def load_blankimage():
    blank_image = np.zeros(shape=(512, 600, 3), dtype = np.int16)
    org_img = blank_image
    return blank_image
```

```python
#ploting image
blank_image=load_blankimage()
plt.imshow(blank_image)
```

```python
#draw a straight line
line_img = load_blankimage()
#cv2.line(image_name, pt1=(x1,y1), pt2=(x2, y2),color=(r,g,b), thickness=px)
cv2.line(line_img, pt1=(0,0), pt2=(600,512), color=(0,0,255),thickness=5)
plt.imshow(line_img)
```

```python
#draw circle
img_circle = load_blankimage()
cv2.circle(img=img_circle, center=(350,350), radius=100, color=(0,255,0), thickness=-1)
plt.imshow(img_circle)
```

```python
#draw a rectangle
# pt1 = top left
# pt2 = bottom right
rect_img = load_blankimage()
cv2.rectangle(img=rect_img, pt1=(200,300), pt2=(450,400), color=(255,0,0), thickness=-1)
plt.imshow(rect_img)
```

```python
#text in the blank_image
text_image = load_blankimage()
cv2.putText(text_image, text = "Mamun", org=(100, 200), fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=4,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(text_image)
```

```python
poly_img=load_blankimage()
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)

cv2.polylines(poly_img,[vertices],isClosed=True,color=(255,255,0),thickness=5)
plt.imshow(poly_img)
```
