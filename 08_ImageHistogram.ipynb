```python
#import packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load image
image_bgr = cv2.imread('images/girl.jpg')
```

```python
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
#working with (b,g,r) image
#hist_values = cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
hist_values = cv2.calcHist([image_bgr],[0],mask=None, histSize=[256],ranges=[0,256])
```

```python
#ploting blue curve
plt.plot(hist_values)
```

```python
#hist_values = cv2.calcHist([image_bgr],[2],mask=None, histSize=[256],ranges=[0,256])
#plt.plot(hist_values)
```

```python
image = image_bgr
Color = ('b', 'g', 'r')
for i, color in enumerate(Color):
    histr = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histr, Color=color)
    plt.xlim([0,256])
plt.title("Girl Image")
plt.show()
    
```

```python
image = image_bgr
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(image,cv2.CV_32F,1,0,ksize=5)
plt.imshow(sobelx,cmap="gray")
```
