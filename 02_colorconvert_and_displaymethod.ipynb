```python
#load the necessary packages
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#method for display image
def display_image(image, cmap=None):
    fig = plt.figure(figsize = (12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap)
```

```python
#method for load image
def load_image():
    image = cv2.imread('images/dog.jpg')
    return image
```

```python
#load image from images folder
#image = cv2.imread('images/dog.jpg')
image = load_image()
original_image = image
```

```python
#display original image
display_image(original_image)
```

```python
#using opencv to display original image
cv2.imshow("image", original_image)
cv2.waitKey(0)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
display_image(image_gray , cmap = "gray")
```

```python
#display gray imageusing opencv
cv2.imshow("gray",image_gray)
cv2.waitKey(0)
```
